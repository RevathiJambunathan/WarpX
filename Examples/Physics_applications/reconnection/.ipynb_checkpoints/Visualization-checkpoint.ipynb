```python
plotfile_path = './diags/'

diag_name = 'diags' # E.g., diagnostics.diags_names = diag
iteration = 0
plotfile = plotfile_path+'/{}{:05d}'.format(diag_name, iteration)
field = 'Ex'
species = 'electron'
ds = yt.load( plotfile ) # Load the plotfile
#from yt.frontends.boxlib.api import AMReXDataset
#ds = AMReXDataset(plotfile)
#ds.field_list # Print all available quantities
# Index of data in the plotfile
ds_index = ds.index
ds.index
# Print the number of grids in the simulation
ds_index.grids.shape
#ds.max_level
# Left and right physical boundary of each grid
ds_index.grid_left_edge
#ds_index.grid_right_edge
# List available fields
ds.field_list
#ds.domain_dimensions
```

### Plot data with yt

```python
iteration = 1
dt_iteration = 1000000000000000000
field='Bz'
while iteration<301:
    plotfile = plotfile_path+'/{}{:05d}'.format(diag_name, iteration)
    ds = yt.load( plotfile ) # Load the plotfile
    sl = yt.SlicePlot(ds,1, field, aspect=.1) # Create a sliceplot object
    #sl.annotate_particles(width=(10.e-6, 'm'), p_size=50, ptype='electron', col='red')
    #sl.annotate_particles(width=(10.e-6, 'm'), p_size=50, ptype='proton', col='blue')
    #sl.set_zlim(field, -100,100)
    sl.annotate_grids() # Show grids
    sl.set_log(field,False)
    sl.show() # Show the plot
    sl.save(plotfile_path+'/'+field+'_'+'{:05d}'.format(iteration)+'.png')
    iteration = iteration + dt_iteration
    #sl.annotate_grids() # Show grids
    #sl.show() # Show the plot
```

### Store quantities in numpy arrays, and plot with matplotlib

```python
# Get field quantities
#all_data_level_0 = ds.covering_grid(level=0,left_edge=ds.domain_left_edge, dims=ds.domain_dimensions)
# store grid number 2 into miy_grid
#plotfile_path = './EffectOfGatherAndDeposBuffer/default_setting_for_buffers/diags/'
diag_name = 'diag1' # E.g., diagnostics.diags_names = diag
iteration = 0
dt_iteration = 5000
while iteration < 161:
    plotfile = plotfile_path+'/{}{:05d}'.format(diag_name, iteration)
    species = 'electron'
    ds = yt.load( plotfile ) # Load the plotfile
    field = 'Ey_fp'
    lev =0
    my_grid = ds.index.grids[lev]
    print(ds.index.grids.shape)
    # Get left and right edges of my_grid
    my_grid.LeftEdge
    print(my_grid.LeftEdge)
    my_grid.RightEdge
    print(my_grid.RightEdge)
    # Get Level of my_grid
    my_grid.Level
    print(my_grid.Level)
    # left edge of the grid, in number of points
    #my_grid.start_index
    my_field = my_grid['raw', field].squeeze().v
    print(my_field.shape)
    f_shape = my_field.shape

    #extent = np.array()
    #[my_grid.LeftEdge[0],my_grid.RightEdge[0],my_grid.LeftEdge[0],my_grid.RightEdge[0],my_grid.LeftEdge[0],my_grid.RightEdge[0]
    # Plot image
    plt.rcParams.update({'font.size': 22})
    plt.figure()
    plt.imshow(my_field[:,f_shape[1]//2,:,0], aspect='auto',cmap='seismic')
    #plt.imshow(my_field[:,:,0,0], aspect='auto',cmap='seismic')
    plt.colorbar()
    plt.ylabel('x')
    plt.xlabel('y')
    plt.tight_layout()
    plt.savefig(plotfile_path+'/'+field+'_'+'{:05d}'.format(iteration)+'_lev'+str(lev)+'.png')
    #plt.scatter(z,x,s=50,c='k')
    
    my_field.max()
    iteration = iteration + dt_iteration
```
