```python
#plotfile_path = './../Injection_p2/InjectionBelowRstar_plus_3dR/plasma_injection/removeparticleinsiderstar_plus_3dR/diags/'
#plotfile_path2 = plotfile_path

#plotfile_path = './noParticleInjection/CorotatingE_3cells_outside_Rstar/'
#plotfile_path = './InjectionBelowRstar/plasma_injection/OnlyTheoryEBAtParticleLocation/diags/'
#plotfile_path = './InjectionBelowRstar/plasma_injection/OnlyTheoryEBAtParticleLocation/withoutmonopoleE/diags/'
#plotfile_path = './InjectionBelowRstar/plasma_injection/SelfconsistentEB_AddMonopole/limit_theta_injection45to65_withmonopole/do_nodal/diags/'
plotfile_path = './diags/'
plotfile_path2 = plotfile_path

#plotfile_path = './InjectionFraction_p4/RampOmega/180k/nx512_dx351/Conductivity_macroscopic/sigma_within_rstar_minus_sigma/diags/'
#plotfile_path2 = plotfile_path


############ Plot comparison of rho and rhoGJ on surface and along equator ############
from yt.units.yt_array import YTQuantity
from yt.units import kboltz
from numpy.random import random
import numpy as np
from yt.utilities.exceptions import YTUnitOperationError


############# Functions needed to compute theoretical rhoGJ ################
def r_cl(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')

    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc))**0.5

    return r

def rad(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')

    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5

    return r

def thetaval(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')

    thetaval = np.arccos((data["z"]-zc)/(data["rad"]))

    return thetaval


def thetadeg(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')

    thetadeg = np.arccos((data["z"]-zc)/(data["rad"])) * 180/3.14

    return thetadeg

def rho_GJ(field,data):
 
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')

    omega = 6245.8
    B = 8.03e-6
    epsilon = 8.854e-12
    c_theta = (data["z"]-zc)/data["rad"]
#    s_theta = data["r_cl"].d/data["rad"].d
#    r_ratio = r_star.d/data["rad"].d
#    c_2theta = c_theta*c_theta - s_theta*s_theta
    r_ratio = r_star.d/data["rad"].d
    c_2theta = np.cos(2.0*data["thetaval"])
    r3 = r_ratio*r_ratio*r_ratio
    rho_GJ = -1.0*epsilon * omega*B*r3*(1. - 3.*c_2theta)

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 10000] = 1.0
    return (rho_GJ*outside.d)

def divE_theory(field,data):
 
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')

    omega = 6245.8
    B = 8.03e-6
    epsilon = 8.854e-12
#    c_theta = (data["z"]-zc)/data["rad"]
#    s_theta = data["r_cl"].d/data["rad"].d
#    r_ratio = r_star.d/data["rad"].d
#    c_2theta = c_theta*c_theta - s_theta*s_theta
    r_ratio = r_star.d/data["rad"].d
    c_2theta = np.cos(data["thetaval"]) * np.cos(data["thetaval"])
    r3 = r_ratio*r_ratio*r_ratio
    rho_GJ = 2.0* omega*B*r3*(1. - 3.*c_2theta)

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 100] = 1.0
    inside_star = np.zeros_like(data["r_cl"])
    inside_star[data["rad"] < 12032 + 3*700] = 1.0
    rho_GJ_out = 2.0* omega*B*r3*r_ratio*(1. - 3.*c_2theta)
    outside_star = np.zeros_like(data["r_cl"])
    outside_star[data["rad"] >= 12032 + 3*700] = 1.0    
    return (rho_GJ*outside.d*inside_star.d + rho_GJ_out*outside_star.d)


def rho2(field,data):
    return data["rho"].d



def Bx_theory(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm') 
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5  
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/(r)
    r3 = r_ratio*r_ratio*r_ratio
        
    Br = 2 * B * r3 * c_theta
    B_theta = B * r3 * s_theta
    Bphi = 0.0
    Bx_ext = Br*c_phi*s_theta + B_theta*c_phi*c_theta    
    return Bx_ext.d


def By_theory(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')  
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5 
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/(r)
    r3 = r_ratio*r_ratio*r_ratio
    
    Br = 2 * B * r3 * c_theta
    B_theta = B * r3 * s_theta
    Bphi = 0.0
    By_ext = Br*s_phi*s_theta + B_theta*s_phi*c_theta    
    return By_ext.d

def Bz_theory(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')   
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5   
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/(r)
    r3 = r_ratio*r_ratio*r_ratio    
    Br = 2 * B * r3 * c_theta
    B_theta = B * r3 * s_theta
    Bphi = 0.0
    Bz_ext = Br*c_theta - B_theta*s_theta
    return Bz_ext.d

def Ex_theory_quad(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')
    dR = 700    

    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    #r = data["rad"]
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/(r)
    r3 = r_ratio*r_ratio*r_ratio

    Er = omega*r_star*B*r3*r_ratio*(1.0-3.0*c_theta*c_theta) 
    #+ (2.0/3.0)*omega*B*r_star*r_ratio*r_ratio
    E_theta = (-1)*omega*B*r_star*r3*r_ratio*(c_theta*s_theta*2.0)
    Ex_ext = Er*c_phi*s_theta + E_theta*c_phi*c_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 12032 + 3.*dR] = 1.0

 
    return outside.d*Ex_ext.d

def Ex_theory_mono(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')
    dR = 350    

    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    #r = data["rad"]
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/(r)
    r3 = r_ratio*r_ratio*r_ratio

    Er = 0. + (2.0/3.0)*omega*B*r_star#*r_ratio*r_ratio
    #E_theta = (-1)*omega*B*r_star*r3*r_ratio*(c_theta*s_theta*2.0)
    Ex_ext = Er*c_phi*s_theta #+ E_theta*c_phi*c_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 12032 + 3.*dR] = 1.0
 
    return outside.d*Ex_ext.d

def Ex_theory_cor(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')
    dR = 700
    omega = 6245.8
    B = 8.03e-6
    r = (((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5)
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio    
    Er = omega*r_star*r3*B*s_theta*s_theta
    E_theta = (-1)*omega*B*r_star*r3*(c_theta*s_theta*2.0)
    Ex_ext = Er*c_phi*s_theta + E_theta*c_phi*c_theta


    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] <= 12032 + 3.*dR] = 1.0

    return outside.d*Ex_ext.d

def Ey_theory_quad(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')
    dR = 700    
    omega = 6245.8
    B = 8.03e-6

    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio
    Er = omega*r_star*B*r3*r_ratio*(1.0-3.0*c_theta*c_theta) 
    #+ (2.0/3.0)*omega*B*r_star*r_ratio*r_ratio
    E_theta = (-1)*omega*B*r_star*r3*r_ratio*(c_theta*s_theta*2.0)
    Ey_ext = Er*s_phi*s_theta + E_theta*s_phi*c_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 12032+3.*dR] = 1.0

 
    return outside.d*Ey_ext.d

def Ey_theory_mono(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    r_star = YTQuantity(12032, 'm')
    dR = 700    
    omega = 6245.8
    B = 8.03e-6

    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio
    Er = 0. + (2.0/3.0)*omega*B*r_star*r_ratio*r_ratio
    #E_theta = (-1)*omega*B*r_star*r3*r_ratio*(c_theta*s_theta*2.0)
    Ey_ext = Er*s_phi*s_theta# + E_theta*s_phi*c_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 12032+3.*dR] = 1.0

    return outside.d*Ey_ext.d

def Ey_theory_cor(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    dR = 350    
    r_star = YTQuantity(12032, 'm')
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio
    Er = omega*r_star*r3*B*s_theta*s_theta
    E_theta = (-1)*omega*B*r_star*r3*(c_theta*s_theta*2.0)
    Ey_ext = Er*s_phi*s_theta + E_theta*s_phi*c_theta
    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] <= 12032 + 3.*dR] = 1.0

 
    return outside.d*Ey_ext.d 

def Ez_theory_quad(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    dR = 700    
    r_star = YTQuantity(12032, 'm')
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio
    Er = omega*r_star*B*r3*r_ratio*(1.0-3.0*c_theta*c_theta) #+ (2.0/3.0)*omega*B*r_star*r_ratio*r_ratio
    E_theta = (-1)*omega*B*r_star*r3*r_ratio*(c_theta*s_theta*2.0)
    Ez_ext = Er*c_theta - E_theta*s_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 12032+3.*dR] = 1.0
 
    return outside.d*Ez_ext.d

def Ez_theory_mono(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    dR = 350   
    r_star = YTQuantity(12032, 'm')
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio
    Er = 0. + (2.0/3.0)*omega*B*r_star*r_ratio*r_ratio
    #E_theta = (-1)*omega*B*r_star*r3*r_ratio*(c_theta*s_theta*2.0)
    Ez_ext = Er*c_theta #- E_theta*s_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 12032+3.*dR] = 1.0
 
    return outside.d*Ez_ext.d

def Ez_theory_cor(field,data):
    xc = YTQuantity(90000, 'm')
    yc = YTQuantity(90000, 'm')
    zc = YTQuantity(90000, 'm')
    dR = 700    
    r_star = YTQuantity(12032, 'm')
    omega = 6245.8
    B = 8.03e-6
    r = ((data["y"]-yc)*(data["y"]-yc) + (data["x"]-xc)*(data["x"]-xc) + (data["z"]-zc)*(data["z"]-zc))**0.5
    c_theta = (data["z"]-zc)/r
    s_theta = data["r_cl"]/r
    c_phi   = (data["x"]-xc)/data["r_cl"]
    s_phi   = (data["y"]-yc)/data["r_cl"]
    r_ratio = r_star/r
    r3 = r_ratio*r_ratio*r_ratio
    Er = omega*r_star*r3*B*s_theta*s_theta
    E_theta = (-1)*omega*B*r_star*r3*(c_theta*s_theta*2.0)
    Ez_ext = Er*c_theta - E_theta*s_theta

    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] <= 12032+3.*dR] = 1.0

 
    return outside.d*Ez_ext.d
    #return outside.d*Er.d   

def Ex_total_ext(field,data):
    tot = data["Ex_theory_quad"] + data["Ex_theory_cor"] + data["Ex_theory_mono"]
    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 300] = 1.0

    return tot*outside.d

def Ey_total_ext(field,data):
    tot = data["Ey_theory_quad"] + data["Ey_theory_cor"] + data["Ey_theory_mono"]
    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 300] = 1.0

    return tot*outside.d

def Ez_total_ext(field,data):
    tot = data["Ez_theory_quad"] + data["Ez_theory_cor"] + data["Ez_theory_mono"]
    outside = np.zeros_like(data["r_cl"])
    outside[data["rad"] > 300] = 1.0


    return tot*outside.d    
    
def EdotB_theory(field,data):
    EdotB = data["Ex_total_ext"]*data["Bx_theory"] + data["Ey_total_ext"]*data["By_theory"] + data["Ez_total_ext"]*data["Bz_theory"]
    return EdotB

def EdotB_sim(field,data):
    EdotB = data["Ex"]*data["Bx"] + data["Ey"]*data["By"] + data["Ez"]*data["Bz"]
    return EdotB.d

def Ex_diff(field,data):
    Exdiff = data["Ex"].d - data["Ex_total_ext"]
    return Exdiff

def Ey_diff(field,data):
    Eydiff = data["Ey"].d - data["Ey_total_ext"]
    return Eydiff

def Ez_diff(field,data):
    Ezdiff = data["Ez"].d - data["Ez_total_ext"]
    return Ezdiff

def Bx_diff(field,data):
    Bxdiff = data["Bx"].d - data["Bx_theory"]
    return Bxdiff

def By_diff(field,data):
    Bydiff = data["By"].d - data["By_theory"]
    return Bydiff

def Bz_diff(field,data):
    Bzdiff = data["Bz"].d - data["Bz_theory"]
    return Bzdiff

def Ex_sim_ext(field,data):
    Ex_sim_ext = data["Ex"].d + data["Ex_theory_mono"]
    return Ex_sim_ext

def Ey_sim_ext(field,data):
    Ey_sim_ext = data["Ey"].d + data["Ey_theory_mono"]
    return Ey_sim_ext

def Ez_sim_ext(field,data):
    Ez_sim_ext = data["Ez"].d + data["Ez_theory_mono"]
    return Ez_sim_ext
```
